```python
#-*- coding: utf-8 -*-
SHOP_URL = "http://www.rocketsalad.co.kr"
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98 Safari/537.36',
}
```

```python
list_html = requests.get(url=SHOP_URL, headers=headers).content
list_lxml = BeautifulSoup(list_html, 'lxml', from_encoding='utf-8')
```

```python
category_list = ['/shop/shopbrand.html?xcode=113&type=X', '/shop/shopbrand.html?xcode=118&type=X', '/shop/shopbrand.html?xcode=115&type=X', '/shop/shopbrand.html?xcode=116&type=X']
```

```python
http://www.rocketsalad.co.kr/shop/shopbrand.html?xcode=113&type=X
```

```python
url_list = []
img_list = []

url_list1 = []
img_list1 = []
name_list = []
for category in category_list:
    list_html = requests.get(url=str(SHOP_URL+category), headers=headers).content
    list_lxml = BeautifulSoup(list_html, 'lxml', from_encoding='utf-8')
    
    url_list += [SHOP_URL+i.findParent('a').get('href') for i in list_lxml.select('span.Tahoma')]
    img_list1 += [SHOP_URL+i.get('src') for i in list_lxml.select('a > img')]
    name_list += [i.find_all(text=True)[0] for i in list_lxml.select('span.Tahoma')]
    
   
    
    for i, img in enumerate(img_list1):
        if img.find('shopimages') == -1:
            del img_list1[i]
        else:
            img_list.append(img)
print(len(url_list))
print(len(img_list))
print(len(name_list))
```
