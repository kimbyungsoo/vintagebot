```python
#-*- coding: utf-8 -*-
SHOP_URL = "http://m.cemeterypark.kr"
ALL_URL = "/product/list.html?cate_no=24"
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98 Safari/537.36'}
```

```python
list_html = requests.get(url=SHOP_URL+ALL_URL, headers=headers).content
list_lxml = BeautifulSoup(list_html, 'lxml', from_encoding='utf-8')
```

```python
url_list = [SHOP_URL+i.get('href') for i in list_lxml.select('div.thumbnail > a')]
img_list = ['http:'+i.get('src') for i in list_lxml.select('div.thumbnail > a > img')]
name_list = [i.find_all(text=True)[0] for i in list_lxml.select('div.information > div.name > a')]
```
